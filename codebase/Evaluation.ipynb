```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import os
```

```python
results_dir = "/content/drive/MyDrive/CS GY 6643: Final Project/YOLOv11/final_eval"

# List all files in the directory
files = [f for f in os.listdir(results_dir) if f.endswith(".csv")]

# Initialize a dictionary to store data for plotting
model_data = {}

# Read each results.csv and extract the mAP50-95 metric
for file in files:
    file_path = os.path.join(results_dir, file)
    try:
        df = pd.read_csv(file_path)
        if "metrics/mAP50-95(B)" in df.columns:
            model_name = os.path.splitext(file)[0]  # Use file name (without extension) as model name
            model_data[model_name] = df["metrics/mAP50-95(B)"].values
        else:
            print(f"Skipping {file}: 'metrics/mAP50-95(B)' column not found.")
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Plot the mAP50-95 for all models
plt.figure(figsize=(12, 6))
for model_name, mAP_values in model_data.items():
    plt.plot(range(1, len(mAP_values) + 1), mAP_values, label=model_name, marker='o')

# Add labels, legend, and title
plt.xlabel("Epochs")
plt.ylabel("mAP50-95")
plt.title("mAP50-95 Comparison Across Models")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the plot
output_plot_path = os.path.join(results_dir, "mAP50-95_comparison.png")
plt.savefig(output_plot_path)
plt.show()

print(f"Plot saved to: {output_plot_path}")
```
